Question:-

->If 20 is the radius of the circle to be drown and origin is its Centre then plot the first octant of the circle.

->If 10 is the radius of the circle to be drown and (4, 4) is its Centre then plot the first and second octant of the circle.

```python
#initial work:
print("Center: ")
x=int(input())
y=int(input())

r=int(input("input redius: "))  #input any redius
ip=[0,r]                        #initial point
dp=1-r                          #initial decision parameter
print (ip,dp)
```

```python
it=0

xk=ip[0]
yk=ip[1]
p=dp

tx=[]
ty=[]
tx.append(xk)
ty.append(yk)
print("iteration", "\t" ,"  xk,yk", "\t\t" ,"pk","\t\t"  ,"xk+1,yk+1\n")
while((xk<yk)):
    print(it, "\t\t  " ,xk,",",yk,"\t\t" ,p, end =" ")
    if(p<0):
        p=p+2*xk+3
        xk+=1
        
    else:
        p=p+2*(xk-yk)+5
        xk+=1
        yk-=1
    
    print("\t\t"  ,xk,",",yk)
    tx.append(xk)
    ty.append(yk)
    it+=1
```

```python
octant=[]

#octant 1
for i,j in zip(tx,ty):
    octant.append([i+x,j+y])
#print(octant)

#octant 2
t1=tx[::-1]
t2=ty[::-1]
for i,j in zip(t1,t2):
    octant.append([j+x,i+y])

#octant 3
t1=np.array(tx)*(-1)
t2=ty
for i,j in zip(t1,t2):
    octant.append([j+x,i+y])

#octant 4
t1=np.array(tx[::-1])
t2=np.array(ty[::-1])*(-1)
for i,j in zip(t1,t2):
    octant.append([i+x,j+y])

#octant 5
t1=np.array(tx)*(-1)
t2=np.array(ty)*(-1)
for i,j in zip(t1,t2):
    octant.append([i+x,j+y])

#octant 6
t1=np.array(tx[::-1])*(-1)
t2=np.array(ty[::-1])*(-1)
for i,j in zip(t1,t2):
    octant.append([j+x,i+y])

#octant 7
t1=np.array(tx)
t2=np.array(ty)*(-1)
for i,j in zip(t1,t2):
    octant.append([j+x,i+y])

#octant 8
t1=np.array(tx[::-1])*(-1)
t2=ty[::-1]
for i,j in zip(t1,t2):
    octant.append([i+x,j+y])
    
octant=np.array(octant)
octant
```

```python
def ploting(x):
    z=[]
    z2=[]
    for i in x:
        z.append(i[0]);
        z2.append(i[1]);
    z.append(z[0])
    z2.append(z2[0])
    plt.scatter(z,z2,color='red')
    plt.plot(z,z2,color='green')
    plt.show()
```
