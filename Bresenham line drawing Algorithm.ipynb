```python
xx=[[1,1],[35,40],[10,20],[6,7],[30,18]]
yy=[[6,7],[43,45],[30,18],[1,1],[10,20]]
```

```python
def cal_m(x,y):
    return (y[1]-x[1])/(y[0]-x[0])
```

```python
def cal_Pk(x,y):
    return 2*(y[1]-x[1])-(y[0]-x[0])
```

```python
if __name__=="__main__":
    cc=0
    for x,y in zip(xx,yy):
        cc+=1
        m=cal_m(x,y)
        pk = cal_Pk(x,y)
        z = []
        z2 = []
        z.append(x[0])
        z2.append(y[1])
        print("Case ",cc,":-",x,y)
        for idx in range(100000):
            if(m<1):
                if(pk<0):
                    pk = pk+2*(y[1]-x[1])
                    x[0]+=1
                else:
                    pk = pk+2*(y[1]-x[1])-2*(y[0]-x[0])
                    x[0]+=1
                    x[1]+=1
            else:
                if(pk<0):
                    pk = pk + 2*(y[0]-x[0])
                    x[0]+=1
                else:
                    pk = pk+2*(y[0]-x[0])-2*(y[1]-x[1])
                    x[0]+=1
                    x[1]+=1
            z.append(x[0])
            z2.append(x[1])
            if(x[0]==y[0] or x[1]==y[1]):
                        break;
        
        plt.scatter(z,z2,color='red')
        plt.plot(z,z2,color='green')
        plt.show()
```
