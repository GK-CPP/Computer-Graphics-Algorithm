```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Questions:-

-> Transform (1, 2), (4, 3) and (3, 5) points using translation and scaling factors 9 along -axis and 6 along -axis and rotate 270 with anticlockwise.

-> Transform with matrix (2, −3) and (−4,10) points using translation and scaling factors 3 along -axis and 4 along -axis and rotate 90 with clockwise.

-> Transform (2, 1), (3, 4) and (5, 3) points using translation and scaling factors 6 along -axis and 9 along -axis.

```python
n=int(input("Number of point:"))
```

```python
point=[]
for i in range(0,n):
    xx=int(input("x axis point:"))
    yy=int(input("y axis point:"))
    point.append([xx,yy])

fx=int(input("x axis factor:"))
fy=int(input("y axis factor:"))
factor=[]
factor.append(fx)
factor.append(fy)
```

```python
def ploting(x):
    z=[]
    z2=[]
    for i in x:
        z.append(i[0]);
        z2.append(i[1]);
    z.append(z[0])
    z2.append(z2[0])
    plt.scatter(z,z2,color='red')
    plt.plot(z,z2,color='green')
    plt.show()
```

# Before(Translation,Scaling & Rotate)

```python
point=np.array(point)
factor=np.array(factor)
ploting(point)
point
```

# Translation

```python
t=[]  
for i in range(0,n):
    t.append(point[i]+factor)
t=np.array(t)
t
```

# After Translation

```python
ploting(t)
```

# Scaling

```python
s=[]
for i in range(0,n):
    s.append(t[i]*factor)
s=np.array(s)
s
```

# After Scaling

```python
ploting(s)
```

# Rotate

```python
r=input("Rotation type (ac\c): ") #given input ac=anti-clockwise or c=clockwise
theta=int(input("Angle:"))
theta=math.radians(theta)

new=[]
for i in range(0,n):
    if(r=="ac"):
        xx=int(t[i][0]*round(math.cos(theta))-t[i][1]*round(math.sin(theta)))
        yy=int(t[i][0]*round(math.sin(theta))+t[i][1]*round(math.cos(theta)))
    else:
        xx=int(t[i][0]*round(math.cos(theta))+t[i][1]*round(math.sin(theta)))
        yy=int(t[i][1]*round(math.cos(theta))-t[i][0]*round(math.sin(theta)))
    new.append([xx,yy]);
new=np.array(new)
new
```

# After Rotate

```python
ploting(new)
```
